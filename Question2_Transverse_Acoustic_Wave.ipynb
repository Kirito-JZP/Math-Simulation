```python
# CS7GV2 Mathematics of Light and Sound
# ZHIPENG JIA
# ID 20327934
# Question 2 Transverse Acoustic Wave

%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

amplitude = 0.2
wave_length = 1.5
time_cycle = 30
omega = 2 * np.pi / time_cycle
k = 2 * np.pi / wave_length
phase = 0

x1 = np.random.uniform(1, 10, 600)
y1 = np.random.uniform(1, 2, 600)
x2 = np.random.uniform(1, 10, 600)
y2 = np.random.uniform(1, 2, 600)

fig, ax = plt.subplots()


def animate(i):
    # Longitudinal wave
    # calculate next x position
    # Wave equation: A*sin(wt-kx)
    y1_next0 = y1 + amplitude * np.sin(omega * i - k * x1)

    # clear and add new points
    fig.clear()
    ax0 = fig.add_subplot(211)
    ax0.axis('off')
    ax0.set_title("Transverse Wave (t)")
    ax0.set_xlim(1, 10)
    ax0.set_ylim(0.5, 2.5)
    ax0.scatter(x1, y1_next0, s=25, c="black", marker="o")
    ax0.scatter(x1[300], y1_next0[300], s=25, c="red", marker="o")
    ax0.scatter(x1[500], y1_next0[500], s=25, c="red", marker="o")

    # Transverse wave
    # calculate next y axis
    y1_next1 = y1 + amplitude * np.sin(omega * (i + 20)- k * x1)

    ax1 = fig.add_subplot(212)
    ax1.axis('off')
    ax1.set_title("Transverse Wave (t+dt)")
    ax1.set_xlim(1, 10)
    ax1.set_ylim(0.5, 2.5)
    ax1.scatter(x1, y1_next1, s=25, c="black", marker="o")
    ax1.scatter(x1[300], y1_next1[300], s=25, c="red", marker="o")
    ax1.scatter(x1[500], y1_next1[500], s=25, c="red", marker="o")


ani = animation.FuncAnimation(fig, animate, interval=20)
plt.show()
```
