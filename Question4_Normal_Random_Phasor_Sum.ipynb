```python
# CS7GV2 Mathematics of Light and Sound
# ZHIPENG JIA
# ID 20327934
# Question 4 Normal-like distribution of resultants from a large number of random phasor sums

import random
import numpy as np
import matplotlib.pyplot as plt

x = []
y = []

for i in range(10000):
    p_x = 0
    p_y = 0
    # Sum of 100 random phasors
    for j in range(100):
        A = random.uniform(-1, 1)
        phi = random.uniform(-2 * np.pi, 2 * np.pi)
        p_x = p_x + A * np.cos(phi)
        p_y = p_y + A * np.sin(phi)
    x.append(p_x)
    y.append(p_y)
plt.scatter(x, y, marker='.', c='black')
plt.show()
```
